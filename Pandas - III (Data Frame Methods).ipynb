```python
import pandas as pd 
import numpy as np 
```

```python
ipl=pd.read_csv('ipl-matches.csv')
```

```python
movies=pd.read_csv('movies.csv')
```

```python
# value_counts(series and dataframe) -- it count the rows 
marks=pd.DataFrame({
    'iq':[100,80,90,90],
    'name':['Aman','Shiva','Aditya','Aditya'],
    'marks':[100,80,95,95]
})

marks.value_counts()
```

```python
# find which player has won most potm --> in qualifiers and finals
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()
```

```python
# Toss decision plot
ipl['TossDecision'].value_counts().plot(kind='pie')
```

```python
# how many matches each team has played 
t1=ipl['Team1'].value_counts()
t2=ipl['Team1'].value_counts()

(t1+t2).sort_values(ascending =False)
```
