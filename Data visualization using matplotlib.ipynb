```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

<h1>creating line charts </h1>

```python
days = ['sun','mon','tue','wed','thu','fri','sat']
temp = [35,40,35,40,42,44,41]

plt.plot(days,temp,marker='o')
plt.xlabel('days')
plt.ylabel('temparature in degree celcius')
plt.title('days vs temprature graph')
plt.grid(True)
plt.show()
```

```python
fruits = ['apple','banana','mango','lichi']
no = [10,25,12,40]

myexplode = [0.1, 0.1 , 0.1 , 0.2]

plt.pie(no,labels=fruits,explode=myexplode)
plt.show()
```

```python
# plotting histogram
df= pd.DataFrame({
    'days':['sun','mon','tue','wed','thu','fri','sat'],
    'week1' : [50,20,10,50,40,30,20],
    'week2' : [50,60,20,30,40,50,60],
    'week3' : [30,20,12,15,22,32,14]
    })
df.plot(kind='hist',fill=False,hatch='o')

plt.show()
```

```python
# scatter plot

discount = np.array([10,20,30,40,50,60,70,80])
sales=np.array([4000,5000,6000,7000,8000,9000,4500,5550])

plt.scatter(x=discount,y=sales)
plt.title('sales vs discount')

plt.xlabel('discount offerd')
plt.ylabel('sales in rs')
plt.show()
```
